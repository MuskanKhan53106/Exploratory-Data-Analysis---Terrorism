# 6. Data Visualization

## 6.1 Number of terrorist activites each year

```python
v=terror_data['iyear'].value_counts()
```

```python
plt.figure(figsize=(20,10))
sns.set(style='ticks')
sns.barplot(x=v.index,y=v,palette='coolwarm')

plt.xticks(rotation=90)

plt.title('terrorist_activites',fontsize=20)
plt.xlabel('year',fontsize=15)
plt.ylabel('count',fontsize=15)
plt.show()
```

**Insight** :

*   From here we can see that most of the attack happen in  and The number of terrorist activity have gone up sharply after 2010



## 6.2 Country affected by terrorism

```python
v=terror_data['country_txt'].value_counts()
```

```python
plt.figure(figsize=(20,60))
sns.barplot(x=v,y=v.index,palette='Set1')
plt.xticks(rotation=90)
plt.title('Country affected by terrorism',fontsize=20)
plt.xlabel('count',fontsize=15)
plt.ylabel('Country',fontsize=15)
plt.show()
```

**Insight** :

*   'Iraq', 'Pakistan', 'Afghanistan', 'India' & 'Colombia' are the countries where most terrorist attack happend





## 6.3 Region affected by terrorism

```python
sns.set(style = 'darkgrid')
ax=sns.countplot(x='region_txt',data=terror_data)
plt.xticks(rotation=90)

total = float(len(terror_data['region_txt']))

for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='right')

plt.title('Distribution of Attacks per Region (1970-2017)', fontsize = 20)
plt.xlabel('Region Name',fontsize = 15)
plt.ylabel('Counts',fontsize = 15)
plt.show()
```

**Insight** :

* Most terrorist attack happened in 'Middle East & North Africa' region and 27.8% show the  all records between 1970 and 2017

## 6.5 City affected by terrorism

## 6.6 Terrorist attack Successed or failed

```python
v=terror_data['success'].value_counts()
fig=plt.gcf()
fig.set_size_inches((10,6))
cmap=plt.get_cmap('Set3')
color=cmap(np.arange(len(v)))
plt.pie(v,labels=['successed','failed'],autopct="%0.02f",colors=color,shadow=True)
plt.title('Terrorist attack Successed or failed (1970-2017)', fontsize = (20))
plt.show()
```

**Insight** :

* From above wee can see that approx 89% terrorist attck successed and 11% failed from 1970 to 2017



## 6.7 Suicide or Not Suicide

```python
v=terror_data['suicide'].value_counts()
fig=plt.gcf()
fig.set_size_inches((10,6))
cmap=plt.get_cmap('Pastel1')
color=cmap(np.arange(len(v)))
plt.pie(v,labels=['suicide','not suicide'],autopct="%0.02f",colors=color,shadow=True)
plt.title('Suiciude or Not Suicide(1970-2017)', fontsize = (20))
plt.show()
```

**Insight** :

*  Aprrox 97% terrorist attack was suicide and 3% was not suicide

## 6.11 Terrorist group name

```python
top_20_gnames = terror_data['gname'].value_counts().nlargest(20).index
filtered_data = terror_data[terror_data['gname'].isin(top_20_gnames)]
plt.figure(figsize=(20,10))
colors = sns.color_palette("Set1")
sns.countplot(x='gname', data=filtered_data,color=colors[0])
plt.xticks(rotation=90)
plt.title('terror group name',fontsize=20)
plt.xlabel('terror group name',fontsize=15)
plt.ylabel('count',fontsize=15)
plt.show()
```

**Insight**

*   'Unknown', 'Taliban', 'Islamic State of Iraq and the Levant (ISIL) are most famous trrrorist group form 1970 to 2017.

## 6.13 Highest number of people killed by country

```python
from itertools import groupby
killed_people=terror_data.groupby(['country_txt'], as_index=False)['nkill'].sum().sort_values(by='nkill',ascending=False)[:50]
plt.figure(figsize=(20,40))
sns.barplot(x='nkill', y='country_txt', data=killed_people, palette='Set1')
plt.title('Highest number of people killed by country',fontsize=20)
plt.xlabel('No. of people killed',fontsize=20)
plt.ylabel('country name',fontsize=20)
plt.xticks(rotation=90)  # Rotate x labels if needed
plt.show()
```

**Insight** :

* Highest number of death of people happend in Iraq, Afganisthan, Pakisthan, Nigeria & India from 1970 to 2017.



## 6.14 Highest No. of people killed by Region

```python
v=terror_data.groupby(['region_txt'], as_index=False)['nkill'].sum().sort_values(by='nkill',ascending=False)
plt.figure(figsize=(20,8))
plt.plot(v['region_txt'], v['nkill'], 'g--')
plt.title('Highest No. of people killed by Region',fontsize=20)
plt.xlabel('Regions',fontsize=20)
plt.ylabel('No. of people killed',fontsize=20)
plt.xticks(rotation=90)  # Rotate x labels if needed
plt.show()
```

**Insight** ;

*  Highest No. of people killed in Middle east and Africa region(1970-2017).



## 6.15 Highest No. of people killed by State

```python
v=terror_data.groupby(['provstate'], as_index=False)['nkill'].sum().sort_values(by='nkill',ascending=False)[:50]
plt.figure(figsize=(20,40))
sns.barplot(x='nkill',y="provstate",data=v,palette='cubehelix')
plt.title('Highest No. of people killed by State',fontsize=20)
plt.xlabel('No. of people killed',fontsize=20)
plt.ylabel('State',fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

**Insight** :

* Highest No. of people killed in 'Baghdad', 'Unknown', 'Nineveh', 'Borno' 7 'Al Anbar' states from(1970-2017).



## 6.16 Highest No. of people killed by city

```python
v=terror_data.groupby(['city'], as_index=False)['nkill'].sum().sort_values(by='nkill',ascending=False)[:50]
plt.figure(figsize=(20,40))
sns.barplot(x='nkill',y='city',data=v,palette='dark:salmon_r')
plt.title('Highest No. of people killed by City',fontsize=20)
plt.xlabel('No. of people killed',fontsize=20)
plt.ylabel('City',fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

**Insight** :



*  Highest No. of people killed in 'Unknown', 'Baghdad', 'Mosul', 'Mogadishu', & 'Karachi', 'New York City' cities from(1970-2017).
